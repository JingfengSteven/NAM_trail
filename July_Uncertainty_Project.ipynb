```python
class MyCredit:
  def __init__ (self,data):
    self.id=hasher(data)
    self.data=data
    self.var=0
    self.value=Model_load_value(data)

  def __lt__ (self, other):
    return self.value > other.value

  def get_id(self):
    return self.id

  def get_data(self):
    return self.data

  def get_var(self):
    return self.var

  def get_value(self):
    return self.value
```

```python
def hasher(data):
  Arr=np.array(data)
  str_arr = np.array_str(Arr)

  str_arr = str_arr.replace('[', '').replace(']', '').replace('\n', '')

  Hash=hash(str_arr)
  return Hash
```

```python
def Model_load_var(X):

  predict1=Model_1(np.array([X]))
  predict2=Model_2(np.array([X]))
  predict3=Model_3(np.array([X]))
  predict4=Model_4(np.array([X]))
  predict5=Model_5(np.array([X]))
  predict6=Model_6(np.array([X]))
  predict7=Model_7(np.array([X]))
  predict8=Model_8(np.array([X]))
  predict9=Model_9(np.array([X]))
  predict10=Model_10(np.array([X]))

  su=[]
  su.append(np.array(predict1))
  su.append(np.array(predict2))
  su.append(np.array(predict3))
  su.append(np.array(predict4))
  su.append(np.array(predict5))
  su.append(np.array(predict6))
  su.append(np.array(predict7))
  su.append(np.array(predict8))
  su.append(np.array(predict9))
  su.append(np.array(predict10))

  su=np.array(su)

  su=su.reshape(10,1)
  vari=np.var(su,axis=0)

  return vari
```

```python
def Model_load_value(X):

  mean=(Model_1(np.array([X]))+Model_2(np.array([X]))+Model_3(np.array([X]))+Model_4(np.array([X]))+Model_5(np.array([X]))+Model_6(np.array([X]))+Model_7(np.array([X]))+Model_8(np.array([X]))+Model_9(np.array([X]))+Model_10(np.array([X])))/10

  return mean
```

```python
def calc_var(data):
  Var=Model_load_var(data)
  return Var
```

```python
def calc_value(data):
  Mean=Model_load_value(data)
  return Mean
```

```python
def find_next1(now):
  data_next=copy.deepcopy(now)
  data_next.data[0]=data_next.data[0]-1

  sign=1
  if(data_next.data[0]<0 or data_next.data[1]<0 or data_next.data[2]<0 or data_next.data[3]<0 or data_next.data[4]<0):
    sign=-1
  data_next.id=hasher(data_next.data)
 # data_next.var=calc_var(data_next.data)
 # data_next.value=calc_value(data_next.data)
  return data_next,sign
```

```python
def find_next2(now):
  data_next=copy.deepcopy(now)

  data_next.data[1]=data_next.data[1]-1

  sign=1
  if(data_next.data[0]<0 or data_next.data[1]<0 or data_next.data[2]<0 or data_next.data[3]<0 or data_next.data[4]<0):
    sign=-1
  data_next.id=hasher(data_next.data)
 # data_next.var=calc_var(data_next.data)
 # data_next.value=calc_value(data_next.data)
  return data_next,sign
```

```python
def find_next4(now):
  data_next=copy.deepcopy(now)

  data_next.data[3]=data_next.data[3]-1

  sign=1
  if(data_next.data[0]<0 or data_next.data[1]<0 or data_next.data[2]<0 or data_next.data[3]<0 or data_next.data[4]<0):
    sign=-1
  data_next.id=hasher(data_next.data)
#  data_next.var=calc_var(data_next.data)
#  data_next.value=calc_value(data_next.data)
  return data_next,sign
```

```python
def find_next5(now):
  data_next=copy.deepcopy(now)

  data_next.data[4]=data_next.data[4]-1
  sign=1
  if(data_next.data[0]<0 or data_next.data[1]<0 or data_next.data[2]<0 or data_next.data[3]<0 or data_next.data[4]<0):
    sign=-1
  data_next.id=hasher(data_next.data)
 # data_next.var=calc_var(data_next.data)
 # data_next.value=calc_value(data_next.data)
  return data_next,sign
```

```python
def check_uncertainty(point,hold,boundary):
  reflec=False
  flag=[]
  now=MyCredit(point)
  que=[]
  heapq.heappush(que, now)
  flag.append(now.id)
  iterm=0
  for i in range(1,len(boundary)):
    bod=MyCredit(boundary[i])
    heapq.heappush(que, bod)
    flag.append(bod.id)

  while(len(que)!=0):
    now=heapq.heappop(que)
    now.var=Model_load_var(now.data)
    iterm+=1
 #   print(now.data[0:5],np.array(now.value),np.array(now.var))
    data1,sign1=find_next1(now)
    data2,sign2=find_next2(now)
    data3,sign3=find_next3(now)
    data4,sign4=find_next4(now)
    data5,sign5=find_next5(now)

    if(data1.id not in flag and sign1==1):
     # data1.var=calc_var(data1.data)
      data1.value=calc_value(data1.data)
      flag.append(data1.id)
      heapq.heappush(que, data1)
    if(data2.id not in flag and sign2==1):
     # data2.var=calc_var(data2.data)
      data2.value=calc_value(data2.data)
      flag.append(data2.id)
      heapq.heappush(que, data2)
    if(data3.id not in flag and sign3==1):
     # data3.var=calc_var(data3.data)
      data3.value=calc_value(data3.data)
      flag.append(data3.id)
      heapq.heappush(que, data3)
    if(data4.id not in flag and sign4==1):
     # data4.var=calc_var(data4.data)
      data4.value=calc_value(data4.data)
      flag.append(data4.id)
      heapq.heappush(que, data4)
    if(data5.id not in flag and sign5==1):
     # data5.var=calc_var(data5.data)
      data5.value=calc_value(data5.data)
      flag.append(data5.id)
      heapq.heappush(que, data5)

    if(now.var<hold):
      reflec=True
      return now,reflec,iterm

  return now,reflec,iterm
```

```python
def find_boundary(check):
  prex1=check[0]
  prex2=check[1]
  prex3=check[2]
  boundary=[]
  #x1<x2<x3
  for x_1 in range(5):
    for x_2 in range(5):
      for x_3 in range(5):
        if(x_3<=prex3 and x_2+x_3<=prex2+prex3 and x_1+x_2+x_3==prex1+prex2+prex3):
          boundary.append([x_1,x_2,x_3,check[3],check[4],check[5],check[6],check[7],check[8],check[9]])
  return boundary
```
